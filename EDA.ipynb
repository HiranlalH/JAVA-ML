```python
# Sales by Category (Bar Charts):

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=cldf)
plt.title('Sales by Category')
plt.show()
```

```python
# Profit by Category (Bar Charts):

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Profit', data=cldf)
plt.title('Profit by Category')
plt.show()
```

```python
#Sales by Region (Pie Chart):
sales_by_region = cldf.groupby('Region')['Sales'].sum()

plt.figure(figsize=(8, 8))
sales_by_region.plot.pie(autopct='%1.1f%%')
plt.title('Sales by Region')
plt.ylabel('')
plt.show()


```

```python
# Distribution of Sales (Histogram)

plt.figure(figsize=(10, 6))
sns.histplot(cldf['Sales'], bins=30, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.show()

```

```python
# Distribution of Profit(Histogram)

plt.figure(figsize=(10, 6))
sns.histplot(cldf['Profit'], bins=30, kde=True)
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.show()

```

```python
# Sales vs. Profit (Scatter Plot)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=cldf)
plt.title('Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
import numpy as np
```
